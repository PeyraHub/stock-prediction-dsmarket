## Holt Winters

```python
# Convertir la columna 'date' a formato datetime si no está en ese formato ya
df_ts['date'] = pd.to_datetime(df_ts['date'])
```

```python
# Crear un diccionario para almacenar los modelos, pronósticos y métricas de rendimiento para cada ID y parámetros
models_dict = {}
forecasts_dict = {}
scores_dict = {}

best_models_dict = {}
best_scores_dict = {}
```

```python
# Definir los parámetros de búsqueda del Holt-Winters
param_grid = {
    'trend': ['add', None], #'mul',
    'seasonal': ['add', None], #'mul', 
    'seasonal_periods': [ 3, 4, 5, 6, 12, 24, 48, 52]  # Puedes agregar más valores si deseas probar diferentes períodos estacionales
}
```

```python
from itertools import product
from datetime import datetime
import os
import numpy as np
import pandas as pd

from darts import TimeSeries
from statsmodels.tsa.holtwinters import ExponentialSmoothing as HWES
from sklearn.metrics import mean_squared_error

best_scores_dict = {}
best_models_dict = {}

# Iterar sobre todas las combinaciones de parámetros
for trend, seasonal, seasonal_periods in product(param_grid['trend'],
                                                 param_grid['seasonal'],
                                                 param_grid['seasonal_periods']):
    for group_id, group_data in df_ts.groupby('id'):
        # Ordenar por fecha por si acaso
        group_data = group_data.sort_values('date').copy()

        # Crear el TimeSeries object para el grupo actual
        time_series = TimeSeries.from_dataframe(
            group_data,
            time_col='date',
            value_cols='units'
        )
        
        # Utilizar las últimas 8 filas para el conjunto de prueba
        train_data = time_series[:-8]
        test_data = time_series[-8:]

        # Evitar log de valores no positivos
        small_number = 0.01
        # En versiones nuevas de Darts usamos .values() en vez de .pd_dataframe()
        train_values = train_data.values().squeeze()
        train_data_np = np.log1p(train_values + small_number)

        # Ajustar el modelo Holt-Winters (statsmodels)
        model = HWES(
            train_data_np,
            trend=trend,
            seasonal=seasonal,
            seasonal_periods=seasonal_periods
        )
        
        # Fit del modelo
        try:
            model_fit = model.fit()
        except Exception as e:
            print(f"Failed to fit model for group {group_id} with parameters: "
                  f"trend={trend}, seasonal={seasonal}, seasonal_periods={seasonal_periods}.")
            print(f"Error: {e}")
            continue
        
        # Predicciones próximas 8 semanas
        try:
            predictions = model_fit.forecast(steps=8)
        except Exception as e:
            print(f"Failed to generate forecasts for group {group_id} with parameters: "
                  f"trend={trend}, seasonal={seasonal}, seasonal_periods={seasonal_periods}.")
            print(f"Error: {e}")
            continue

        # Inversa del log1p
        predictions = np.expm1(predictions)

        # Convertir test_data a NumPy array para la métrica (también con .values())
        test_data_np = test_data.values().squeeze()
        
        # MSE y RMSE
        mse = mean_squared_error(test_data_np, predictions)
        rmse = np.sqrt(mse)
        
        # Actualizar mejor modelo por grupo
        if group_id not in best_scores_dict or rmse < best_scores_dict[group_id]:
            best_scores_dict[group_id] = rmse
            best_models_dict[group_id] = {
                'model': model_fit,
                'trend': trend,
                'seasonal': seasonal,
                'seasonal_periods': seasonal_periods,
                'predictions': predictions
            }

# ---- Guardado de modelos y predicciones ----

today_date = datetime.today().strftime('%Y-%m-%d')

models_folder = f'models_{today_date}_2weeks'
predictions_folder = f'predictions_{today_date}_2weeks'

os.makedirs(models_folder, exist_ok=True)
os.makedirs(predictions_folder, exist_ok=True)

for group_id, model_info in best_models_dict.items():
    model_filename = (f'model_best_trend_{model_info["trend"]}'
                      f'_seasonal_{model_info["seasonal"]}'
                      f'_periods_{model_info["seasonal_periods"]}'
                      f'_ID_{group_id}.pkl')
    prediction_filename = (f'predictions_best_trend_{model_info["trend"]}'
                           f'_seasonal_{model_info["seasonal"]}'
                           f'_periods_{model_info["seasonal_periods"]}'
                           f'_ID_{group_id}.csv')
    
    # Guardar modelo
    model_info['model'].save(os.path.join(models_folder, model_filename))
    # Guardar predicciones
    pd.DataFrame(model_info['predictions'], columns=['forecast']).to_csv(
        os.path.join(predictions_folder, prediction_filename),
        index=False
    )
```

```python
def calculate_mape(y_true, y_pred):
    """
    Calcula el Mean Absolute Percentage Error (MAPE) entre las predicciones y los valores reales.

    Args:
        y_true (array-like): Array o lista de los valores reales.
        y_pred (array-like): Array o lista de las predicciones del modelo.

    Returns:
        float: El valor del MAPE.
    """
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
# Crear un DataFrame para almacenar las métricas de rendimiento
metrics_df = pd.DataFrame(columns=[
    'ID', 
    'Trend', 
    'Seasonal', 
    'Seasonal_Periods', 
    'RMSE', 
    'MAPE'
])

# Lista donde iremos acumulando cada fila
rows = []

# Guardar las métricas del RMSE y MAPE en la lista de diccionarios
for group_id, model_info in best_models_dict.items():
    row = {
        'ID': group_id,
        'Trend': model_info["trend"],
        'Seasonal': model_info["seasonal"],
        'Seasonal_Periods': model_info["seasonal_periods"],
        'RMSE': best_scores_dict[group_id],
        'MAPE': calculate_mape(test_data_np, model_info['predictions'])
    }
    rows.append(row)

# Convertir la lista en un DataFrame
metrics_df = pd.DataFrame(rows)

# Guardar el DataFrame de métricas en un archivo CSV
metrics_df.to_csv(f'metrics_{today_date}_2weeks.csv', index=False)
```

```python
# Obtén una lista de todos los archivos csv en el directorio
all_files = glob.glob('./predictions_2023-12-03_8weeks/*.csv')  # Reemplaza con la ruta a tu carpeta

# Crea una lista de dataframes
list_df = []
```

```python
for filename in all_files:
    # Obtiene el ID del nombre del archivo
    id = filename.split('_ID_')[1].split('.csv')[0]

    # Selecciona la serie de tiempo correspondiente al ID del DataFrame original
    ts = df_ts[df_ts['id'] == id]

    # Encuentra la fecha final de las predicciones
    end_date = ts['date'].iloc[-1]
    
    # Calcula la fecha de inicio restando 8-7=7 semanas a la fecha final
    start_date = end_date - timedelta(weeks=7)

    # número de semanas entre las fechas
    num_weeks = (end_date - start_date).days // 7

    # lista de todas las fechas en el rango
    date_range = [start_date + timedelta(weeks=x) for x in range(num_weeks+1)]

    # Lee el archivo csv actual
    df = pd.read_csv(filename, index_col=None, header=0)

    # Agrega la columna de fechas al DataFrame
    df['date'] = date_range

    # Agrega la columna de id al DataFrame
    df['id'] = id

    # Agrega el DataFrame a la lista
    list_df.append(df)
```
