```python
import pandas as pd

INPUT_PATH = '/Users/anapeyra/Desktop/NUCLIO/Entregables/TFM/stock-prediction-dsmarket/data_dsmarket/modified/sales_calendar_prices_full.csv'
OUTPUT_PATH = '/Users/anapeyra/Desktop/NUCLIO/Entregables/TFM/stock-prediction-dsmarket/data_dsmarket/interim/full_df_gb_date.csv'

def main():
    chunksize = 100_000  # puedes subir/bajar esto si quieres
    first_chunk = True
    offset = 0  # para la columna Unnamed: 0

    # Borramos cualquier fichero anterior
    # (si existe no pasa nada, lo sobreescribimos en el primer .to_csv)
    print("Empezando construcción de full_df_gb_date.csv en chunks...")

    for i, chunk in enumerate(pd.read_csv(INPUT_PATH, chunksize=chunksize)):
        print(f"Procesando chunk {i} (filas {len(chunk)})...")

        # Asegurar tipo fecha
        chunk["date"] = pd.to_datetime(chunk["date"])

        # Columnas temporales
        iso = chunk["date"].dt.isocalendar()
        chunk["year"] = iso.year
        chunk["week"] = iso.week
        chunk["year_week"] = (
            chunk["year"].astype(str)
            + "-"
            + chunk["week"].astype(str).str.zfill(2)
        )
        chunk["month"] = chunk["date"].dt.month_name()

        # revenue
        chunk["revenue"] = chunk["units"] * chunk["sell_price"]

        # Seleccionar columnas finales
        out = chunk[
            [
                "id", "item", "category", "department", "store", "store_code",
                "region", "date", "year_week", "month", "year",
                "units", "sell_price", "revenue",
            ]
        ].copy()

        # Crear columna Unnamed: 0 global (no solo dentro del chunk)
        out.insert(0, "Unnamed: 0", range(offset, offset + len(out)))
        offset += len(out)

        # Escribir al CSV final (primero con cabecera, luego sin)
        mode = "w" if first_chunk else "a"
        header = first_chunk
        out.to_csv(OUTPUT_PATH, mode=mode, header=header, index=False)
        first_chunk = False

    print("Terminado. Fichero guardado en:", OUTPUT_PATH)

if __name__ == "__main__":
    main()
```
